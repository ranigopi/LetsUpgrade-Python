```
 num = int(input("enter a number:"))
 
if num < 0:
   print("Enter a positive number")
else:
   sum = 0
 
   while(num > 0):
       sum += num
       num -= 1
   print("The sum is", sum)
```

```
 num = int(input("enter a number: "))
 
for i in range(2, num):
    if num % i  == 0:
        print("not prime number")
        break
else:
    print("prime number")
```
