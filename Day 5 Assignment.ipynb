```
list1=[10,20,40,60,70,80]
list2=[5,15,25,35,45,60]
size1 = len(list1) 
 
size2 = len(list2) 
 
  
 
res = [] 
 
i, j = 0, 0
 
  
 
while i < size1 and j < size2: 
 
    if list1[i] < list2[j]: 
 
      res.append(list1[i]) 
 
      i += 1
 
  
 
    else: 
 
      res.append(list2[j]) 
 
      j += 1
 
  
 
res = res + list1[i:] + list2[j:]
 
print ("merge of list is ",res)
```
